### Object Oriented Programming

 class

```python
class human:
    pass

person = human()
person.name = "Shashi"
person.surname = "Kumar"
person.year_of_birth = 1998


print(person)
print("%s %s was born in %d." %
      (person.name, person.surname, person.year_of_birth))
```

```python
class human:
    def __init__(self, name, surname, year_of_birth):
        self.name = name
        self.surname = surname
        self.year_of_birth = year_of_birth
```

 Methods

```python
class human:
    def __init__(self, name, surname, year_of_birth):
        self.name = name
        self.surname = surname
        self.year_of_birth = year_of_birth

    def age(self, current_year):
        return current_year - self.year_of_birth

    def __str__(self):
        return "%s %s was born in %d ." % (self.name, self.surname, self.year_of_birth)

person = human("shashi", "kumar", 1998)
print(person)
print(person.age(2019))
```

Inheritance

```python
class Student(human):
    def __init__(self, student_id, *args, **kwargs):
        super(Student, self).__init__(*args, **kwargs)
        self._student_id = student_id

charlie = Student(1, 'amaravathi', 'B', 2006)
print(charlie)
print(type(charlie))
print(isinstance(charlie, human))
print(isinstance(charlie, object))
```

```python
class Student(human):
    def __init__(self, student_id, *args, **kwargs):
        super(Student, self).__init__(*args, **kwargs)
        self._student_id = student_id

    def __str__(self):
        return super(Student, self).__str__() + " And has ID: %d" % self._student_id

shashi = Student(1, 'shashi', 'kumar', 2006)
print(shashi)
```

Encapsulation

```python
def summer(a, b):
    return a + b

print(summer(1, 1))
print(summer(["a", "b", "c"], ["d", "e"]))
print(summer("abra", "cadabra"))
```

# Files


```python
# Open the text.txt we made earlier
with open("superman.txt","w") as file:
  my_file = file.write("good morning,myself superman i am going to save world from supervillan")
```

```python
# Now read again
with open('superman.txt', 'r') as my_file:
  print(my_file.read())
```

```python
# Readlines returns a list of the lines in the file.
with open('superman.txt', 'r') as my_file:
  my_file.readlines()
```

### Writing to a File


```python
# Add the second argument to the function, 'w' which stands for write
my_file = open('superman.txt','w+')
```

```python
# Write to the file
my_file.write('This is a new line')
```

```python
# Seek to the start of file (index 0)
my_file.seek(0)
```

```python
# Read the file
my_file.read()
```
